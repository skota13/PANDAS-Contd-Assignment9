# PANDAS Contd Assignment 9

```python
import pandas as pd
df1 = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/06_Stats/US_Baby_Names/US_Baby_Names_right.csv')
```

# Delete unnamed columns

```python
df2 = df1.drop(['Unnamed: 0'], axis=1)
df2.head(5)
```

# Show the distribution of male and female

```python
df2[['Gender','Count','State']].groupby(['Gender']).describe().transpose()
```

# Show the top 5 most preferred names

```python
import numpy as np
top5=df2[['Name','Count']].groupby('Name').max()
top5.sort_values('Count',ascending=0).head(n=5)
```

# What is the median name occurence in the dataset?

```python
df2.loc[df2['Count']==df2['Count'].median()].head(5)
```

# Distribution of male and female born count by states

```python
df2[['Gender','State','Count']].groupby(['Gender','State']).describe().transpose()
```
